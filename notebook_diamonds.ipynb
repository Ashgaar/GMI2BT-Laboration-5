```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = "./diamonds.csv"
data = pd.read_csv(path)
```

```python
data.isnull().any()
```

```python
del data[data.columns[0]]
data
```

```python
data.describe().T
```

```python
data.dtypes
```

```python
data.groupby(['cut','color'])['price'].mean()
```

```python
plt.boxplot(data['carat'])
plt.ylabel('Carat')
plt.title('Boxplot of carat')
```

```python
sns.violinplot(x='clarity', y='price', data=data)
plt.show()
```

```python
sns.catplot(x='clarity', y='price', data=data, hue = 'cut', kind = 'bar')
plt.show()
```

```python
sns.violinplot(x='clarity', y='carat', data=data)
plt.show()
```

```python
sns.jointplot(x='carat', y='price', data=data)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Correlation Map')
ax=sns.heatmap(data.corr(), linewidth=3.1, annot=True, center=1)
```

```python
sns.lmplot(x='carat', y='price', hue = 'color', data=data)
```

```python
sns.regplot(x='carat', y='price', data=data)
plt.show()
```
